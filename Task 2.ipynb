# Writing a Simulation
Implement in R or Python a solution to the following problem

_Imagine there is a country in which couples only want boys.
Couples continue to have children until they have their first boy. If they get a boy, they stop getting children.
What is the long-term ratio of boys to girls in the country?_

```r
# the problem can be solved using geometric distribution

# theoretically
p = 0.5
print(paste("Probability of having a girl:", p))
exp = p/(1-p)   
print(paste("Expected number of girls:", exp))
print(paste("As we know the expected number of boys is 1, the long-term ratio is 1:", exp))
```

```r
# simulation
set.seed(42)
n = seq(100, 10000, 100)
exp = 0

# generate increasingly large samples from a geometric distribution with 
# propability of success in each trial of 0.5 
for(i in 1:length(n)){
  sim = rgeom(n[i], 0.5)
  exp[i] = mean(sim)
  #print(exp[i])
}

# plot the differences between the sample means and the theoretical mean of 1
dif = abs(1-exp)
plot(n, dif)
lines(spline(n, dif))
# expected value approaches the theoretical value 1 (for number of girls) 
# with increasing sample size
```
